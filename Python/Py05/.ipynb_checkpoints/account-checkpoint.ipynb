```python

    a.account() 계좌생성 메서드 (은행명, 계좌주, 계좌번호)
    만든 계좌수 정보 메서드
    deposit 입금 메서드
    withdraw 출금 메서드
    정보 출력 메서드 잔고 계좌 번호 예금주 은행
    이자 지급 //입금 5회시 1프로
    여러 객체 생성 3개 이상 리스트에 저장기능 append()
    객체 순회 리스트로 100만 이상 계좌 찾기
    입금 출금 내역 조회 기능 업데이트
```

```python
import random

class Account():
    account_made_num = 0
    
    def __init__(self): ##계좌 정보 초기값및 양식 저장         
        self.bank_name = "SC은행"
        self.account_name = "홍길동"
        self.acoount_num = "111-11-111111"
        self.account_pound_sum = "0"
        
class Bank_made_account(Account): ##계좌생성메서드 이름과 잔액을 넣어 인스턴스 생성 + 계좌번호 + 계좌생성횟수 기록
        def __init__(self, name, pound):
            super().__init__()
            self.account_name = name
            self.account_pound_sum = pound
                    
            i = random.randint(100,999)
            d = random.randint(10,99)
            e = random.randint(100000,999999)
    
            self.account_num = f"{i}-{d}-{e}"
            print("계좌가 생성되었습니다.")
            
            self.deposit_history = []
            self.withdraw_history = []
            
            Account.account_made_num += 1
            print(f"은행이름:{self.bank_name} 계좌번호:{self.account_num} 예금주 이름:{self.account_name} 잔액:{self.account_pound_sum:,}")

        def get_account_num(self):
            print(f"SC은행 총계좌수 :{Account.account_made_num}")

        def deposit(self):
            
            while True:
                try:
                    depo = int(input("입금액을 입력해주세요 (1원 이상만 가능): "))
                    if depo >= 1:
                        self.account_pound_sum += depo
                        self.deposit_history.append(depo)
                        print(f"{depo:,}원 입금 성공! 잔액: {self.account_pound_sum:,}원")
                                                
                        if len(self.deposit_history) % 5 == 0:  ##입금 횟수에 따른 이자 지급
                            interest = int(self.account_pound_sum * 0.01)
                            self.account_pound_sum += interest
                            print(f"{len(self.deposit_history)}회 입금 달성! 이자 {interest:,}원 지급!")
                            print(f"이자 적용 후 잔액: {self.account_pound_sum:,}원")

                        break
           
                    else:
                        print("1원 이상만 입금 가능합니다.")
                except ValueError:
                    print("숫자를 입력해주세요.")

        def withdraw(self):
           
            print(f"현재 잔액은 {self.account_pound_sum:,}원입니다.")
     
            if self.account_pound_sum <= 0:
                print("잔액이 부족해 출금이 불가능합니다.")
                return
     
            while True:
                try:
                    wi = int(input("출금 액수를 입력해주세요: "))
                    if wi <= 0:
                        print("1원 이상만 출금 가능합니다.")
                    elif wi > self.account_pound_sum:
                        print("잔액이 모자랍니다.")
                    else:
                        self.account_pound_sum -= wi
                        self.withdraw_history.append(wi)
                        print(f"{wi:,}원 출금 완료! 잔액: {self.account_pound_sum:,}원")
                        break
                except ValueError:
                    print("숫자를 입력해주세요.")

        
        def display_info(self):
           
            print(f"은행이름:{self.bank_name}계좌번호:{self.account_num} 예금주 이름:{self.account_name} 잔액:{self.account_pound_sum:,}")
            print("입금 내역")
            for om in range(len(self.deposit_history)):
                print("  ",om+1,self.deposit_history[om])
                print(f"잔고 : {self.account_pound_sum:,}")
            print("출금 내역")
            for om in range(len(self.withdraw_history)):
                print("  ",om+1,self.withdraw_history[om])
                print(f"잔고 : {self.account_pound_sum:,}")
                
        def bank_run(self):
            print("은행업무를 도와드리겠습니다.")
        
            while True:
                try:
                    choice = int(input("\n1번: 입금 | 2번: 출금 | 3번: 계좌정보 조회 | 4번: 종료\n선택: "))
                    if choice == 1:
                        self.deposit()
                    elif choice == 2:
                        self.withdraw()
                    elif choice == 3:
                        self.display_info()
                    elif choice == 4:
                        print("이용해주셔서 감사합니다.")
                        break
                    else:
                        print("1~4 중에서 선택해주세요.")
                except ValueError:
                    print("숫자를 입력해주세요.")
```

```python
Account()
my_account_1 = Bank_made_account("쿠퍼플래그", 1000000)
my_account_2 = Bank_made_account("르브론제임스", 40000000000)
my_account_3 = Bank_made_account("돈치치", 150000000)
my_account_2.get_account_num()
```

```python
my_account_1.bank_run()
```

```python
my_account_2.bank_run()
```

```python
my_account_4 = Bank_made_account("웨스트브록",500000)
account_list = [my_account_1, my_account_2, my_account_3, my_account_4]

print("잔액 1,000,000원 이상 고객 목록")
for account in account_list:
    if account.account_pound_sum >= 1000000:
        account.display_info()
```
